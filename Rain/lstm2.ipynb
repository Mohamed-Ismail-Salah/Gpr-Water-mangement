```python
y_train= dataset[['rainfall']]
X_train= dataset.drop(labels=['rainfall'], axis=1)

print('X_train has shape:',X_train.shape,'\ny_train has shape:',y_train.shape)
```

```python
X_train=np.array(X_train)
y_train=np.array(y_train)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train,y_train, test_size=0.20, random_state=42)
```

```python
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.fit_transform(X_test)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_train.shape
```

```python
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
X_test.shape
```

```python
from keras.layers import LSTM
regressor = Sequential()

regressor.add(LSTM(units = 256, return_sequences = True, input_shape = (X_train.shape[1], 1)))
#regressor.add(Dense(units = 64, activation='relu'))

regressor.add(Dropout(0.25))




regressor.add(LSTM(units = 128, return_sequences = True))
#regressor.add(Dense(units = 64, activation='relu'))

regressor.add(Dropout(0.25))


regressor.add(LSTM(units = 128, return_sequences = True))
#regressor.add(Dense(units = 64, activation='relu'))

regressor.add(Dropout(0.25))

regressor.add(LSTM(units = 128))
#regressor.add(Dense(units = 32, activation='relu'))

#regressor.add(Dropout(0.25))




regressor.add(Dense(1, activation='sigmoid'))
```

```python
# defining loss function, optimizer, metrics and then compiling model
regressor.compile(loss="binary_crossentropy", optimizer='adam', metrics=['accuracy'])
```

```python
# summary of model layers
regressor.summary()
```

```python
# training the model on training dataset
history = regressor.fit(X_train, y_train, epochs=100, batch_size=500,validation_split=0.2)
```

```python
# predicting target attribute on testing dataset
score =regressor.evaluate(X_test, y_test, verbose=1)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
  
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history)

pd.DataFrame(history.history)[["accuracy", "val_accuracy"]].plot()
plt.title("Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("accuracy")

pd.DataFrame(history.history)[["loss", "val_loss"]].plot()
plt.title("Model Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
```
