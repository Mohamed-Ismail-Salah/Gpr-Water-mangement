```python
dataset = pd.read_csv('../input/rainfall-prediction/dataset1.csv')
```

```python
# remove attribute 'day'
dataset.drop(['day'],axis=1,inplace=True)
dataset.shape
```

```python
##dataset.label.value_counts()
```

```python
dataset = pd.get_dummies(dataset,columns=['rainfall'],prefix="",prefix_sep="")  
dataset
```

```python
y_train_multi= dataset[['yes','no']]
X_train_multi= dataset.drop(labels=['yes','no'], axis=1)

print('X_train has shape:',X_train_multi.shape,'\ny_train has shape:',y_train_multi.shape)
```

```python
from sklearn.preprocessing import LabelBinarizer

y_train_multi = LabelBinarizer().fit_transform(y_train_multi)
y_train_multi 
```

```python
X_train_multi=np.array(X_train_multi)
y_train_multi=np.array(y_train_multi)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_multi,y_train_multi, test_size=0.20, random_state=42)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_train.shape
```

```python
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
X_test.shape
```

```python
 dataset.info()
```

```python
from keras.layers import LSTM
regressor = Sequential()

regressor.add(LSTM(units = 128, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 128, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))

regressor.add(Dense(units = 2))
```

```python
# defining loss function, optimizer, metrics and then compiling model
regressor.compile(loss="binary_crossentropy", optimizer='adam', metrics=['accuracy'])
```

```python
# summary of model layers
regressor.summary()
```

```python



history =regressor.fit(X_train , y_train, epochs=50, batch_size=32, validation_split=0.1 )
```

```python
# predicting target attribute on testing dataset
score =regressor.evaluate(X_test, y_test, verbose=1)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
  
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history)

pd.DataFrame(history.history)[["accuracy", "val_accuracy"]].plot()
plt.title("Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("accuracy")

pd.DataFrame(history.history)[["loss", "val_loss"]].plot()
plt.title("Model Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
```
